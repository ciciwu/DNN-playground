```python
import torch
import torchvision
from torch.utils.data import Dataset, DataLoader
import numpy as np
import math
```

```python
np.loadtxt('./wine.data', delimiter=',', dtype=np.float32,skiprows=1)
```

```python
class WineDataset(Dataset):
  def __init__(self, transform):
    self.xy = np.loadtxt('./wine.data', delimiter=',', dtype=np.float32,skiprows=1)
    # self.x = torch.tensor(xy[:,1:], dtype=torch.float32)
    # self.y = torch.tensor(xy[:,[0]], dtype=torch.float32)
    self.n_samples = self.xy.shape[0]
    self.transform = transform

  def __getitem__(self,index):
    #return self.x[index], self.y[index]
    sample =  self.xy[index,1:],  self.xy[index,0]
    return self.transform(sample)

  def __len__(self):
    return self.n_samples


```

```python
class ToTensor():
  def __call__(self,sample):
    inputs,target = sample
    target = target-1
    target = torch.tensor(target, dtype=torch.long)
    return (torch.tensor(inputs, dtype=torch.float32),
           target)
```

```python
ds = WineDataset(transform=ToTensor())
```

```python
dataloader=DataLoader(dataset=ds, batch_size=64,shuffle=True)
```

```python
import torch.nn as nn
```

```python
from torch.optim import Adam
from torch.nn import MSELoss
```

```python
class model(nn.Module):
  def __init__(self, n_feature):
    super().__init__()
    self.net = nn.Sequential(
    nn.Linear(n_feature,20),
    nn.ReLU(),
    nn.Linear(20,3),
    #nn.Softmax(dim=1)
    )

    #self.relu = nn.ReLU()

  def forward(self,x):
    x = self.net(x)
    return x
```

```python
m = model(13)
```

```python
lr = 5e-4
n_epoch = 50_000
```

```python

loss =nn.CrossEntropyLoss()
optimizer=Adam(m.parameters(),lr = lr)
```

```python
total_loss=0
for epoch in range(n_epoch):
  for i,(inputs,label) in enumerate(dataloader):
    #if (i+1)%5==0:

    optimizer.zero_grad()
    pred = m(inputs)

    l = loss(pred,label)
    l.backward()
    optimizer.step()
    total_loss += l.item()


    if (epoch) % 100 == 0:
        print (f'Epoch [{epoch}/{n_epoch}],  Loss: {l.item():.12f}')
```
